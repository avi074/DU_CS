```octave
img = imread('Test_images/baboon.png');
i1 = imresize(img, [256,256]);
imshow(img);
```

```octave
imshow(i1);
```

```octave
y_img = rgb2ycbcr(img);
size(y_img)
```

```octave
[m,n] = size(y_img); 
A = zeros(size(y_img));
B = A;
```

```octave
Q = [16 11 10 16 24 40 51 61 ;
     12 12 14 19 26 28 60 55 ;
     14 13 16 24 40 57 69 56 ;
     14 17 22 29 51 87 80 62 ;
     18 22 37 56 68 109 103 77 ;
     24 35 55 64 81 104 113 92 ;
     49 64 78 87 103 121 120 101;
     72 92 95 98 112 100 103 99];


for channel = 1:3
     for i = 1:8:m-7
          for j = 1:8:n
               sam = y_img(i:i+7, j:j+7, channel);
               freq = imdct(imdct(sam).').';
               freq = Q.*round(freq./Q);
               A(i:i+7, j:j+7, channel) = freq;
               B(i:i+7, j:j+7, channel) = imidct(imidct(freq).').';
          end
     end
end     
```
